```python
import pandas as pd


# Load cleaned IoT data
df = pd.read_csv("C:/Users/david/Desktop/Lerra/mo-it-148/Week 7 Homework/cleaned_iot_data.csv")


# Display first few rows to verify data
print(df.head())
```

```python
# To ensure proper plotting, convert the timestamp column into a datetime format.
# Convert timestamp column to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])

print("Data types after conversion:")
print(df)
```

```python
# Import the required libraries:
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visualization style.
sns.set(style="whitegrid")

# Create the line plot.
plt.figure(figsize=(12, 6))  # Adjust figure size
sns.lineplot(x=df["timestamp"], y=df["temperature"], hue=df["timestamp"], marker="o")


# Rotate x-axis labels for better readability
plt.xticks(rotation=45)


# Add title and labels
plt.title("IoT Sensor Readings Over Time", fontsize=14)
plt.xlabel("Timestamp", fontsize=12)
plt.ylabel("Sensor Value", fontsize=12)


# Show legend
plt.legend(title="Sensor Type")


# Display the plot
plt.show()
```
